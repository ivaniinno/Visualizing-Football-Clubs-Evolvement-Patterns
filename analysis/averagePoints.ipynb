```python
team_info = pd.read_json("../parsing/averagePoints/sorted_teams.json")
```

```python
team_info.drop(["Country_Name", "Link_to_team", "Page"], axis=1, inplace=True)
```

```python
team_info.head()
```

```python
average_points_with_names_df = average_points_df.merge(team_info, how="inner", left_on="TeamID", right_on="TeamID")
```

```python
average_points_with_names_df.head()
```

```python
average_points_with_names_df = average_points_with_names_df.sort_values(by=["TeamID", "Year"])
```

```python
assert average_points_with_names_df.shape[0] == average_points_df.shape[0], "DataFrames have different number of rows"
```

```python
teams_info_json = []

current_team_id = average_points_with_names_df.iloc[0]["TeamID"]
current_team_name = average_points_with_names_df.iloc[0]["Team_name"]
current_team_points = []

for row in average_points_with_names_df.itertuples():
    if row[1] != current_team_id:
        teams_info_json.append({
            "TeamID": current_team_id,
            "TeamName": current_team_name,
            "AveragePoints": current_team_points
        })
        current_team_id = row[1]
        current_team_name = row[4]
        current_team_points = [row[3]]
    else:
        current_team_points.append(row[3])
    
```

```python
teams_info_json.append({
    "TeamID": current_team_id,
    "TeamName": row[4],
    "AveragePoints": current_team_points
})
```
